1. What is hypothesis testing in statistics?
-->Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It helps determine whether there is enough evidence to support a specific claim or hypothesis about a population parameter.

Steps in Hypothesis Testing:
Formulate Hypotheses:

Null Hypothesis (H‚ÇÄ): Assumes no effect or no difference (status quo).

Alternative Hypothesis (H‚ÇÅ or Ha): Represents what we are trying to prove (a significant effect or difference).

Set the Significance Level (Œ±):

Typically, Œ± = 0.05 (5%) is chosen, meaning we accept a 5% risk of rejecting the null hypothesis when it is true.

Select the Appropriate Test:

Based on the data type and distribution, tests like t-test, chi-square test, ANOVA, or z-test are used.

Compute the Test Statistic:

The test statistic (like t-score or z-score) is calculated from the sample data to compare against a critical value.

Make a Decision:

Compare the p-value (probability of obtaining the observed results under H‚ÇÄ) with Œ±.

If p-value ‚â§ Œ±: Reject H‚ÇÄ (significant evidence for H‚ÇÅ).

If p-value > Œ±: Fail to reject H‚ÇÄ (not enough evidence for H‚ÇÅ).

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
-->The null hypothesis (H‚ÇÄ) and the alternative hypothesis (H‚ÇÅ or Ha) are the two key components of hypothesis testing in statistics.

1. Null Hypothesis (H‚ÇÄ)
It represents the default assumption or the status quo.

It states that there is no effect, no difference, or no relationship in the population.

The goal of hypothesis testing is to find evidence to reject H‚ÇÄ.

‚úÖ Example:
A company claims that its new weight-loss supplement has no effect.

H‚ÇÄ: The supplement has no impact on weight loss.

2. Alternative Hypothesis (H‚ÇÅ or Ha)
It is the opposite of H‚ÇÄ and represents what we want to prove.

It states that there is an effect, difference, or relationship.

If strong evidence is found, we reject H‚ÇÄ in favor of H‚ÇÅ.

‚úÖ Example:
If researchers believe the supplement does help in weight loss, then:

H‚ÇÅ: The supplement leads to weight loss.

3. What is the significance level in hypothesis testing, and why is it important?
-->Significance Level (Œ±) in Hypothesis Testing
The significance level (Œ±) is the threshold used to determine whether to reject the null hypothesis (H‚ÇÄ) in hypothesis testing. It represents the probability of making a Type I error, which occurs when we reject a true null hypothesis.

Why is Significance Level Important?
Controls the Probability of False Positives (Type I Error)

A lower Œ± reduces the chances of mistakenly rejecting H‚ÇÄ when it is actually true.

Determines Decision Making

If the p-value (calculated probability) is less than or equal to Œ±, we reject H‚ÇÄ.

If the p-value is greater than Œ±, we fail to reject H‚ÇÄ.

Balances Accuracy & Sensitivity

A very low Œ± (e.g., 0.01) makes the test more strict, reducing false positives but increasing the chance of missing a real effect (Type II error).

A higher Œ± (e.g., 0.10) makes it easier to detect real effects but increases false positives.

4. What does a P-value represent in hypothesis testing?
-->What is a P-Value in Hypothesis Testing?
The p-value (probability value) is a measure of how much evidence we have against the null hypothesis (H‚ÇÄ). It tells us the probability of obtaining a test result at least as extreme as the one observed, assuming that H‚ÇÄ is true.

How to Interpret the P-Value?
Small P-Value (‚â§ Œ±) ‚Üí Reject H‚ÇÄ

The alternative hypothesis (H‚ÇÅ) is likely true.

Large P-Value (> Œ±) ‚Üí Fail to Reject H‚ÇÄ

There is not enough evidence to reject H‚ÇÄ.

The result could have occurred by random chance.

6. What are Type 1 and Type 2 errors in hypothesis testing?
-->Type I and Type II Errors in Hypothesis Testing
When performing hypothesis testing, we make decisions based on sample data, which can sometimes lead to errors. These errors are classified as Type I and Type II errors.

1. Type I Error (False Positive)
Occurs when we reject a true null hypothesis (H‚ÇÄ).

It means we conclude there is an effect when there isn't one.

The probability of making a Type I error is Œ± (significance level).

üìå Example:
A company tests whether a new drug reduces blood pressure.

H‚ÇÄ: The drug has no effect.

H‚ÇÅ: The drug lowers blood pressure.

If we reject H‚ÇÄ when it is actually true, we conclude the drug works when it doesn‚Äôt.

üîπ Consequences: Wasting resources on an ineffective drug.

2. Type II Error (False Negative)
Occurs when we fail to reject a false null hypothesis (H‚ÇÄ).

It means we conclude there is no effect when there actually is one.

The probability of making a Type II error is Œ≤ (related to test power, 1 - Œ≤).

üìå Example:
A manufacturing company tests if a new machine produces fewer defective products.

H‚ÇÄ: The new machine is not better.

H‚ÇÅ: The new machine produces fewer defects.

If we fail to reject H‚ÇÄ when it is false, we miss out on an improvement.

üîπ Consequences: Losing the opportunity to adopt a better process.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
-->One-Tailed vs. Two-Tailed Test in Hypothesis Testing
In hypothesis testing, the choice between a one-tailed and a two-tailed test depends on the research question and the direction of the effect being tested.

1. One-Tailed Test (Directional Test)
Tests for an effect in only one direction (either greater than or less than).

The alternative hypothesis (H‚ÇÅ) specifies whether the parameter is greater or smaller than the null hypothesis (H‚ÇÄ).

The rejection region is only on one side of the distribution.

Examples of One-Tailed Tests:
‚úÖ Right-Tailed Test (H‚ÇÅ: Œº > Œº‚ÇÄ)

Example: Checking if a new fertilizer increases plant growth.

H‚ÇÄ: The fertilizer has no effect.

H‚ÇÅ: The fertilizer increases plant growth.

‚úÖ Left-Tailed Test (H‚ÇÅ: Œº < Œº‚ÇÄ)

Example: Testing if a diet reduces blood sugar levels.

H‚ÇÄ: The diet has no effect.

H‚ÇÅ: The diet reduces blood sugar.

üìå Advantage: More statistical power (higher chance of detecting an effect if it exists).
üìå Disadvantage: If the effect is in the opposite direction, we miss it.

2. Two-Tailed Test (Non-Directional Test)
Tests for an effect in both directions (greater or smaller).

The alternative hypothesis (H‚ÇÅ) states that the parameter is different from the null hypothesis (H‚ÇÄ) but does not specify a direction.

The rejection region is on both sides of the distribution.

Example of a Two-Tailed Test:
Checking if a new medicine has an effect (either increasing or decreasing blood pressure).

H‚ÇÄ: The medicine has no effect on blood pressure.

H‚ÇÅ: The medicine changes blood pressure (could increase or decrease).

üìå Advantage: Detects any significant difference.
üìå Disadvantage: Requires a larger sample size to detect the effect

```python
#practical
#1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results ?
import numpy as np
from scipy import stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test.
    :param sample: List or array of sample data
    :param pop_mean: Known population mean
    :param pop_std: Known population standard deviation
    :param alpha: Significance level (default 0.05)
    :return: Z-score, p-value, and interpretation
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    standard_error = pop_std / np.sqrt(sample_size)

    # Calculate Z-score
    z_score = (sample_mean - pop_mean) / standard_error

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Interpret the results
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")

# Example usage
sample_data = [52, 48, 50, 53, 47, 49, 51, 52, 50, 48]
population_mean = 50
population_std = 3
z_test(sample_data, population_mean, population_std)
```

```python
#26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import chi2_contingency, chisquare, chi2, f_oneway, levene, shapiro, ttest_ind, f
from statsmodels.stats.proportion import proportions_ztest

def check_anova_assumptions(*groups):
    """
    Check assumptions for ANOVA: Normality, Independence, and Equal Variance.
    """
    print("Checking ANOVA assumptions:")

    # Normality test (Shapiro-Wilk test)
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        print(f"Group {i+1} Normality Test: W={stat:.4f}, P-value={p:.4f}")
        if p < 0.05:
            print(f"Group {i+1} is NOT normally distributed.")
        else:
            print(f"Group {i+1} appears to be normally distributed.")

    # Equal Variance Test (Levene's test)
    stat, p = levene(*groups)
    print(f"Levene's Test for Equal Variance: W={stat:.4f}, P-value={p:.4f}")
    if p < 0.05:
        print("Groups do NOT have equal variances (heteroscedasticity detected).")
    else:
        print("Groups have approximately equal variances.")

def one_way_anova(*groups):
    """
    Perform a one-way ANOVA test to compare means between multiple groups and visualize the results.
    """
    check_anova_assumptions(*groups)

    f_stat, p_value = f_oneway(*groups)
    print(f"ANOVA F-statistic: {f_stat:.4f}, P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Reject the null hypothesis: At least one group mean is significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference among group means.")

    # Boxplot visualization
    plt.figure(figsize=(8, 5))
    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title("One-Way ANOVA Test: Comparison of Group Means")
    plt.ylabel("Values")
    plt.show()

def perform_f_test(sample1, sample2):
    """
    Perform an F-test to compare the variances of two datasets and visualize the results.
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    f_stat = var1 / var2
    dfn = len(sample1) - 1  # Degrees of freedom numerator
    dfd = len(sample2) - 1  # Degrees of freedom denominator
    p_value = 1 - f.cdf(f_stat, dfn, dfd)

    print(f"F-Test Statistic: {f_stat:.4f}, P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Reject the null hypothesis: The variances of the two groups are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in variances.")

    # Visualization
    plt.figure(figsize=(8, 5))
    plt.hist(sample1, alpha=0.5, label="Sample 1", bins=15, edgecolor="black")
    plt.hist(sample2, alpha=0.5, label="Sample 2", bins=15, edgecolor="black")
    plt.title("Comparison of Variances (F-Test)")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

def simulate_and_test():
    """
    Simulate data from a normal distribution and perform hypothesis testing.
    """
    np.random.seed(42)

    # Simulate random normal data
    sample1 = np.random.normal(loc=50, scale=5, size=30)
    sample2 = np.random.normal(loc=52, scale=5, size=30)

    print("\nIndependent T-Test Results:")
    perform_t_test(sample1, sample2)

    print("\nOne-Way ANOVA Test Results:")
    one_way_anova(sample1, sample2)

    print("\nF-Test for Comparing Variances:")
    perform_f_test(sample1, sample2)

    print("\nChi-Square Test for Population Variance:")
    perform_variance_test(sample1, pop_variance=25)  # Assuming population variance = 25

    print("\nZ-Test for Proportions:")
    perform_proportion_z_test(count1=50, nobs1=200, count2=60, nobs2=220)  # Example proportions test

simulate_and_test()
```
